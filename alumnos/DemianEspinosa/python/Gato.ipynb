## Tablero

La clase Board crea una matriz de 3x3 para representar la cuadrícula del juego de Gato y contiene varios métodos que resultarán útiles.

```python
class Board():
    def __init__(self):
        self.b = [['' for x in range(3)] for x in range(3)]
 
    #Este método determina si hay un empate o un ganador en el tablero actual.
    def winner(self):
        for p in ['X', 'O']:
            for r in range(3):
                if all(x == p for x in self.b[r]):
                    return p
 
            for c in range(3):
                if all(r[c] == p for r in self.b):
                    return p
 
            if self.b[0][0] == p and self.b[1][1] == p and self.b[2][2] == p:
                return p
 
            if self.b[0][2] == p and self.b[1][1] == p and self.b[2][0] == p:
                return p
 
        return 'T' if self.full() else ''
 
    def flat(self):
        return flatten(self.b)
 
    #Este método checa si el tablero está lleno.
    def full(self):
        return not any(x == '' for x in self.flat())
 
    #Este método sirve para contar los turnos transcurridos en el tablero.
    def count(self):
        return sum(int(x != '') for x in self.flat())
     
    '''Este método "hashea" el tablero y permite utilizarlo como una llave.
        Lo utilizamos en un diccionario en un método posterior.'''
    def __hash__(self):
        return ''.join(self.flat()).__hash__()
 
    #Este método sirve para comparar con otros tableros y así poder utilizar los tableros en árboles.
    def __cmp__(self, other):
        return self.flat() < other.flat()
 
    #Este método devuelve una de las coordenadas vacías de manera aleatoria.
    def randomEmptyCoord(self):
        if self.full():
            raise ValueError('No empty spots')
        while True:
            r, c = random.randint(0, 2), random.randint(0, 2)
            if self.b[r][c] == '':
                return Coord(r, c)
 
    #Este método marca la coordenada (r,c) con el caracter dado como parámetro.
    def set(self, r, c, m):
        if self.b[r][c] != '':
            raise Exception('wat')
        self.b[r][c] = m
     
    #Este método devuelve el caracter en la coordenada (r,c).
    def get(self, r, c):
        return self.b[r][c]
     
    #Este método dibuja el tablero en su estado actual.
    def draw(self):
        for r in range(3):
            for c in range(3):
                print (self.b[r][c] if self.b[r][c] != '' else ' '),
            print
```

## Jugador

## Juego

La clase Game es la que simula el juego. Los parámetros son los dos jugadores que van a participar y el estado del tablero si es que el juego ya está empezado.

```python
class Game:
    #Este constructor le dice al jugador A que utiizará las Xs y al jugador B los Os.
    def __init__(self, playerA, playerB, startBoard = None):
        self.history = []
 
        self.board = startBoard or Board()
 
        playerA.mark = 'X'
        playerB.mark = 'O'
 
        self.players = [playerA, playerB]
     
    '''Este método corre el juego. Mientras siga el juego, selecciona por turnos al jugador
        actual para que haga su movimiento. Se actualiza el tablero en cada turno.'''
    def playGame(self):
        t = self.board.count()
 
        while self.board.winner() == '':
            p = self.players[t % 2].play(self.board, self.history)
            self.history.append(Move(p, self.players[t % 2].mark))
            self.board.set(p.r, p.c, self.players[t % 2].mark)
            t += 1
 
        return self.board.winner()
```

## BelowPlayer

La clase BelowPlayer representa al jugador que tira abajo del jugador anterior.

```python
class BelowPlayer(Player):
    def __str__(self):
        return 'BelowPlayer'
    
    #Si no se ha tirado o no existe el "abajo" del último movimiento, tira al azar.
    def play(self, board, history):
        if len(history) != 0:
            p = history[-1].pos
        else:
            p = board.randomEmptyCoord()
 
        if board.get((p.r + 2) % 3, p.c) == '':
            return Coord((p.r + 2) % 3, p.c)
        else:
            return board.randomEmptyCoord()
```

## RandomPlayer

La clase RandomPlayer representa al jugador que tirará de manera aleatoria. Contiene el método simulateGame que se utilizará para el jugador de tipo PureMonteCarlo.

```python
class RandomPlayer(Player):
    def __str__(self):
        return 'RandomPlayer'
 
    def play(self, board, history):
        return board.randomEmptyCoord()
 
    #Este método simula un juego con un tablero dado y dos jugadores aleatorios. Regresa al ganador.
    @staticmethod
    def simulateGame(board, m, mark):
        board_ = copy.deepcopy(board)
        board_.set(m.r, m.c, mark)
        game = Game(RandomPlayer(), RandomPlayer(), startBoard=board_)
        return game.playGame()
```

## Pure Monte Carlo Player

La clase PureMonteCarloPlayer representa al jugador que juega con el algoritmo Monte Carlo tradicional. En este caso para cada turno del jugador, simula una cantidad finita(SIMULATION_LIMIT) de veces un juego entre dos jugadores aleatorios en cada lugar que pueda tirar. De esta manera tira en la posición que mejor resultados le entregue.

```python
class PureMonteCarloPlayer(Player):
    def __str__(self):
        return 'PureMonteCarloPlayer'
 
    def play(self, board, history):
        bestSoFar, bestWins = board.randomEmptyCoord(), 0
 
        for r in range(3):
            for c in range(3):
                if not board.get(r, c): #Para cada tiro posible.
                    cnt = 0
 
                    for k in range(SIMULATION_LIMIT): #Simula juegos aleatorios.
                        if RandomPlayer.simulateGame(board, Coord(r,c), self.mark) == self.mark:
                            cnt += 1
 
                    if cnt > bestWins: #Escoge el mejor.
                        bestSoFar = Coord(r, c)
                        bestWins = cnt
 
        return bestSoFar
```

## MonteCarloTreeSearchPlayer

La clase MonteCarloTreeSearchPlayer representa al jugador que utiliza el algoritmo de búsqueda Monte Carlo Tree Search (MCTS). El algoritmo es el explicado en Wikipedia. Las funciones definidas son exactmente las de la definición del jugador.

```python
MCTSMove = namedtuple('MCTSMove', 'wins plays')

class MonteCarloTreeSearchPlayer(Player):
    def __str__(self):
        return 'MonteCarloTreeSearchPlayer'
     
    #Este método es exactamente la función que está descrita en Wikipedia.
    @staticmethod
    def moveValue(m, total_simulations):
        exploration_parameter = math.sqrt(2)
        return float(m.wins) / m.plays +\
               exploration_parameter * math.sqrt(math.log(total_simulations) / m.plays)
 
    def __init__(self):
        self.tree = dict()
 
    #Este método simula el juego. Se explicará por pasos.
    def play(self, board, history):
        if self.tree.has_key(board): #Si es una jugada ya antes realizada.
            moves = self.tree[board] #Ya tienes movimientos simulados
        else:
            moves = dict() #En caso contrario, inicializa los movimientos posibles
            for r in range(3):
                for c in range(3):
                    if not board.get(r, c): #Para cada casilla no utilizada simula jugar ahí una vez
                        won = RandomPlayer.simulateGame(board, Coord(r,c), self.mark) == self.mark
                        moves[Coord(r, c)] = MCTSMove(int(won), 1)
 
        total_simulations = sum([x.plays for x in moves.values()])
 
        #Simulamos y escogemos el que maximice la función moveValue y actualizamos
        for k in range(SIMULATION_LIMIT):
            m, stats = max(moves.items(),
                           key = lambda (_,x): self.__class__.moveValue(x, total_simulations + k))
 
            won = RandomPlayer.simulateGame(board, m, self.mark) == self.mark
 
            moves[m] = MCTSMove(stats.wins + won, stats.plays + 1)
 
        #Al contrario de lo que nos diría la intuición, se elige el movimiento más simulado.
        m, _ = max(moves.items(), key = lambda (_,x): x.plays)
 
        return m
```

## Gráfica de diferencias

Gráfica de diferencias entre resultados de acuerdo al número de partidas entre dos jugadores.

```python
def plotDif(results, X, O, prepare):
    time = np.arange(len(results) + 1)
    dif = [0]
    
    #Guarda la diferencia de las victorias utilizando un barrido.
    for i in range(len(results)):
        if results[i] == 'T':
            dif.append(dif[i])
        elif results[i] == 'X':
            dif.append(dif[i] + 1)
        else:
            dif.append(dif[i] - 1)
            
    dif = np.array(dif)
    
    prepare()
    plt.plot(time, dif) 
```

## Gráfica de Torta

Gráfica de torta sobre los resultados del total de juegos entre dos jugadores.

```python
def plotTorta(results, X, O, prepare):
    A = B = T = 0
    
    #Guarda la cantidad de veces que gana cada jugador y el total de empates.
    for i in range(len(results)):
        if results[i] == 'T':
            T += 1
        elif results[i] == 'X':
            A += 1
        else:
            B += 1
 
    labels = X, O, 'Ties'
    fracs = [A, B, T]
 
    prepare()
    plt.pie(fracs, autopct='%1.1f%%', startangle=90)
```

Con el siguiente método se representan las gráficas de las 16 parejas ordenadas posibles. 
Cada gráfica representa los juegos y resultados entre el jugador de la fila contra el jugador de la columna, donde el jugador de las filas juega el primero.

```python
def plotf(f, results):
    P = len(results)
    plt.figure(figsize=(12, 12))
    
    for r in range(P):
        for c in range(P):
            rec, a, b = results[r][c]
            
            def p():
                ax = plt.subplot(P, P, r*P + c + 1)
                plt.ylabel(a if c == 0 else '')
                plt.xlabel(b if r == 0 else '')
                ax.xaxis.set_label_position('top')
                
            f(rec, a, b, p)
            
    plt.show()
```

## Conclusiones

Lo que hace el método de Minimax es que para cada jugador garantiza lo mínimo que puedes asegurar ganar y el caso de juego de gato lo mínimo que puedes asegurar es el empate. Como el espacio de búsqueda del juego de Gato es muy pequeño(menor a 3⁹) utilizar el método de Minimax puede garantizar la estrategia óptima para cada jugador sin tener que utilizar mucho tiempo de procesamiento.

Lo que hace el algoritmo MCTS es que en lugar de simular todo el espacio de búsqueda, simula una cantidad fija de movidas aleatorias que se podrían dar y memoriza si fue una movida deseable. Es muy similar al Monte Carlo Puro, pero en este caso va memorizando las jugadas posibles en cada juego que se ha jugado. 

Es por esto que se puede observar en los resultados que el MCTS va a terminar al menos empatando con el Monte Carlo Puro después haber aprendido lo suficiente, y el Minimax por probar todo siempre va a desde el inicio al menos empatar con cualqueira de los dos. Al comparar el MCTS contra el Minimax, podemos observar que en un espacio de búsqueda grande, conviene usar el MCTS porque el método Minimax no va a tener tiempo de checar todas las posibilidades. Para un juego tan pequeño como el Gato, lo más recomendable sería usar el método de Minimax ya que asegura la mejor movida en un tiempo razonable. Ejemplos de donde sería mucho más recomendable usar el MCTS por la amplitud del espacio de búsqueda es por ejemplo el juego de ajedrez y el juego de Go.
