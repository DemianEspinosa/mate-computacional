```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
 
%matplotlib inline
 
import numpy as np
import matplotlib.pyplot as plt
import math
import random
import copy
import pdb
from collections import *
 
SIMULATION_LIMIT = 20
N = 10  # how many games to play
 
Coord = namedtuple('Coord', 'r c')
Move = namedtuple('Move', 'pos player')
MCTSMove = namedtuple('MCTSMove', 'wins plays')
 
 
def flatten(x):
    return [inner for outer in x for inner in outer]
 
 
class Board():
    def __init__(self):
        self.b = [['' for x in range(3)] for x in range(3)]
 
    def winner(self):
        # yes, this is lazy.
        for p in ['X', 'O']:
            for r in range(3):
                if all(x == p for x in self.b[r]):
                    return p
 
            for c in range(3):
                if all(r[c] == p for r in self.b):
                    return p
 
            if self.b[0][0] == p and self.b[1][1] == p and self.b[2][2] == p:
                return p
 
            if self.b[0][2] == p and self.b[1][1] == p and self.b[2][0] == p:
                return p
 
        return 'T' if self.full() else ''
 
    def flat(self):
        return flatten(self.b)
 
    def full(self):
        return not any(x == '' for x in self.flat())
 
    def count(self):
        return sum(int(x != '') for x in self.flat())
 
    def __hash__(self):
        return ''.join(self.flat()).__hash__()
 
    def __cmp__(self, other):
        return self.flat() < other.flat()
 
    def randomEmptyCoord(self):
        if self.full():
            raise ValueError('No empty spots')
        while True:
            r, c = random.randint(0, 2), random.randint(0, 2)
            if self.b[r][c] == '':
                return Coord(r, c)
 
    def set(self, r, c, m):
        if self.b[r][c] != '':
            raise Exception('wat')
        self.b[r][c] = m
 
    def get(self, r, c):
        return self.b[r][c]
 
    def draw(self):
        for r in range(3):
            for c in range(3):
                print (self.b[r][c] if self.b[r][c] != '' else ' '),
            print
 
 
class Player:
    def __init__(self):
        self.mark = 'W'
 
    def play(self, board, history):
        pass
 
 
class BelowPlayer(Player):
    def __str__(self):
        return 'BelowPlayer'
 
    # If the position below is available, play there.
    # Otherwise, play a random move.
    def play(self, board, history):
        if len(history) != 0:
            p = history[-1].pos
        else:
            p = board.randomEmptyCoord()
 
        if board.get((p.r + 2) % 3, p.c) == '':
            return Coord((p.r + 2) % 3, p.c)
        else:
            return board.randomEmptyCoord()
 
 
class RandomPlayer(Player):
    def __str__(self):
        return 'RandomPlayer'
 
    def play(self, board, history):
        return board.randomEmptyCoord()
 
    @staticmethod
    def simulateGame(board, m, mark):
        board_ = copy.deepcopy(board)
        board_.set(m.r, m.c, mark)
        game = Game(RandomPlayer(), RandomPlayer(), startBoard=board_)
        return game.playGame()
 
class PureMonteCarloPlayer(Player):
    def __str__(self):
        return 'PureMonteCarloPlayer'
 
    def play(self, board, history):
        bestSoFar, bestWins = board.randomEmptyCoord(), 0
 
        for r in range(3):
            for c in range(3):
                if not board.get(r, c):
                    cnt = 0
 
                    for k in range(SIMULATION_LIMIT):
                        if RandomPlayer.simulateGame(board, Coord(r,c), self.mark) == self.mark:
                            cnt += 1
 
                    if cnt > bestWins:
                        bestSoFar = Coord(r, c)
                        bestWins = cnt
 
        return bestSoFar
 
 
class MonteCarloTreeSearchPlayer(Player):
    def __str__(self):
        return 'MonteCarloTreeSearchPlayer'
 
    @staticmethod
    def moveValue(m, total_simulations):
        exploration_parameter = math.sqrt(2)
        return float(m.wins) / m.plays +\
               exploration_parameter * math.sqrt(math.log(total_simulations) / m.plays)
 
    def __init__(self):
        self.tree = dict()
        # FIXME: call super init...
 
    def play(self, board, history):
        bestSoFar, bestWins = board.randomEmptyCoord(), 0
 
        if self.tree.has_key(board):
            moves = self.tree[board]
        else:
            moves = dict()
            for r in range(3):
                for c in range(3):
                    if not board.get(r, c):
                        won = RandomPlayer.simulateGame(board, Coord(r,c), self.mark) == self.mark
                        moves[Coord(r, c)] = MCTSMove(int(won), 1)
 
        total_simulations = sum([x.plays for x in moves.values()])
 
        for k in range(SIMULATION_LIMIT):
            m, stats = max(moves.items(),
                           key = lambda (_,x): self.__class__.moveValue(x, total_simulations + k))
 
            won = RandomPlayer.simulateGame(board, m, self.mark) == self.mark
 
            stats = MCTSMove(stats.wins + won, stats.plays + 1)
            moves[m] = stats
 
        m, _ = max(moves.items(), key = lambda (_,x): x.plays)
 
        return m
 
 
class Game:
    def __init__(self, playerA, playerB, startBoard=None):
        self.history = []
 
        self.board = startBoard or Board()
 
        playerA.mark = 'X'
        playerB.mark = 'O'
 
        self.players = [playerA, playerB]
 
    def playGame(self):
        t = self.board.count()
 
        while self.board.winner() == '':
            p = self.players[t % 2].play(self.board, self.history)
            self.history.append(Move(p, self.players[t % 2].mark))
            self.board.set(p.r, p.c, self.players[t % 2].mark)
            t += 1
 
        return self.board.winner()
 
 
def plotDif(results, X, O):
    plt.figure(figsize=(6, 6))
    time = np.arange(len(results) + 1)
    dif = [0]
    for i in range(len(results)):
        if results[i] == 'T':
            dif.append(dif[i])
        elif results[i] == 'X':
            dif.append(dif[i] + 1)
        else:
            dif.append(dif[i] - 1)
    dif = np.array(dif)
 
    plt.plot(time, dif)
    plt.xlabel('Number of match')
    plt.ylabel('Difference between players')
    plt.title('Difference between ' + X +  ' respect  ' + O)
    plt.show()
 
 
# plotDif('XXXOTXOTOXOOTOXOOOXOTOXOOTOXOTOXOTOOOOOOOTTTTTTTTTXOXOX')
 
def plotPie(results, X, O):
    A = B = T = 0
    for i in range(len(results)):
        if results[i] == 'T':
            T += 1
        elif results[i] == 'X':
            A += 1
        else:
            B += 1
 
    #make a square figure and axes
    plt.figure(1, figsize=(6, 6))
    ax = plt.axes([0.1, 0.1, 0.8, 0.8])
 
    # The slices will be ordered and plotted counter-clockwise.
    labels = X, O, 'Ties'
    fracs = [A, B, T]
    plt.pie(fracs, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
 
    plt.title('Final count', bbox={'facecolor': '0.8', 'pad': 5})
    plt.show()
 
#plotPie('XXXOTXOTTTTTOXOOTOXOOOXOTTTTTTOXOOTOXOTOXOTOOOOOOOTTTTTTTTTX')
 
 
# Example game.
print 'Example game:'
game = Game(RandomPlayer(), BelowPlayer())
print game.playGame(), "wins!"
game.board.draw()
print
 

 
playerClasses = [BelowPlayer, RandomPlayer, PureMonteCarloPlayer, MonteCarloTreeSearchPlayer]
 
for x in playerClasses:
    for o in playerClasses:
        # Start with a fresh instance every time.
        X = x()
        O = o()
        x_wins = 0
        o_wins = 0
        ties = 0
        record = ""
 
        for k in range(N):
            game = Game(X, O)
            winner = game.playGame()
            record += winner
            if winner == "X":
                x_wins += 1
            elif winner == "O":
                o_wins += 1
            else:
                ties += 1
 
        """
        print X, "vs", O, ":", x_wins, "-", o_wins, "-", ties
        print record
        print
        """
        plotDif(record, str(X), str(O))
        plotPie(record, str(X), str(O))
```
